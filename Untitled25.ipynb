```python


import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt



X = torch.tensor([
    [1200, 3, 20],
    [1500, 4, 15],
    [1800, 3, 18],
    [2000, 5, 10],
    [2200, 4, 8],
    [2500, 5, 5],
    [2700, 4, 7],
    [3000, 5, 2],
    [3200, 6, 1],
    [3500, 5, 3]
], dtype=torch.float32)

# House Prices
y = torch.tensor([
    [200],
    [250],
    [280],
    [350],
    [400],
    [450],
    [500],
    [550],
    [600],
    [650]
], dtype=torch.float32)


X_mean = X.mean(dim=0)
X_std = X.std(dim=0)

X = (X - X_mean) / X_std


X_train = X[:8]
y_train = y[:8]

X_test = X[8:]
y_test = y[8:]

print("Training Data Shape:", X_train.shape)
print("Testing Data Shape :", X_test.shape)



class HousePriceMLP(nn.Module):

    def __init__(self):

        super(HousePriceMLP, self).__init__()

        self.fc1 = nn.Linear(3, 16)

        self.bn1 = nn.BatchNorm1d(16)

        self.fc2 = nn.Linear(16, 8)

        self.fc3 = nn.Linear(8, 1)

        self.relu = nn.ReLU()

        self.dropout = nn.Dropout(0.2)

    def forward(self, x):

        x = self.fc1(x)
        x = self.bn1(x)
        x = self.relu(x)

        x = self.dropout(x)

        x = self.fc2(x)
        x = self.relu(x)

        x = self.fc3(x)

        return x



model = HousePriceMLP()

print("\nModel Architecture:\n")
print(model)



criterion = nn.MSELoss()

optimizer = optim.Adam(model.parameters(), lr=0.01)



epochs = 200

train_losses = []
test_losses = []

print("\nTraining Started...\n")

for epoch in range(epochs):


    model.train()


    predictions = model(X_train)

    train_loss = criterion(predictions, y_train)

    optimizer.zero_grad()

    train_loss.backward()

    optimizer.step()


    model.eval()

    with torch.no_grad():

        test_predictions = model(X_test)

        test_loss = criterion(test_predictions, y_test)

    train_losses.append(train_loss.item())
    test_losses.append(test_loss.item())


    if (epoch + 1) % 20 == 0:

        print(f"Epoch [{epoch+1}/{epochs}]")
        print(f"Train Loss: {train_loss.item():.4f}")
        print(f"Test Loss : {test_loss.item():.4f}")
        print("-" * 40)



model.eval()

with torch.no_grad():

    final_predictions = model(X_test)

    final_mse = criterion(final_predictions, y_test)

print("\n==============================")
print("Final Evaluation")
print("==============================")

print("Final MSE:", final_mse.item())



print("\nPredictions vs Actual Values\n")

for i in range(len(X_test)):

    predicted = final_predictions[i].item()
    actual = y_test[i].item()

    print(f"Predicted Price: {predicted:.2f}")
    print(f"Actual Price   : {actual:.2f}")
    print("-" * 30)



print("\n===================================")
print("Experiment 1 -> Different Learning Rate")
print("Learning Rate = 0.001")
print("===================================")

experiment_model = HousePriceMLP()

experiment_optimizer = optim.Adam(
    experiment_model.parameters(),
    lr=0.001
)

for epoch in range(100):

    experiment_model.train()

    pred = experiment_model(X_train)

    loss = criterion(pred, y_train)

    experiment_optimizer.zero_grad()

    loss.backward()

    experiment_optimizer.step()

print("Final Loss with lr=0.001:", loss.item())



class TanhMLP(nn.Module):

    def __init__(self):

        super(TanhMLP, self).__init__()

        self.fc1 = nn.Linear(3, 16)
        self.fc2 = nn.Linear(16, 8)
        self.fc3 = nn.Linear(8, 1)

        # Tanh Activation
        self.tanh = nn.Tanh()

    def forward(self, x):

        x = self.tanh(self.fc1(x))
        x = self.tanh(self.fc2(x))
        x = self.fc3(x)

        return x

print("\n===================================")
print("Experiment 2 -> Tanh Activation")
print("===================================")

tanh_model = TanhMLP()

tanh_optimizer = optim.Adam(
    tanh_model.parameters(),
    lr=0.01
)

for epoch in range(100):

    tanh_model.train()

    pred = tanh_model(X_train)

    loss = criterion(pred, y_train)

    tanh_optimizer.zero_grad()

    loss.backward()

    tanh_optimizer.step()

print("Final Loss using Tanh:", loss.item())



plt.figure(figsize=(8,5))

plt.plot(train_losses, label="Training Loss")

plt.plot(test_losses, label="Validation Loss")

plt.xlabel("Epochs")
plt.ylabel("Loss")

plt.title("Training vs Validation Loss")

plt.legend()

plt.grid(True)

plt.show()



print("\n===================================")
print("Experiments Comparison")
print("===================================")

print("1) Original Model -> ReLU + lr=0.01")
print("2) Experiment 1   -> ReLU + lr=0.001")
print("3) Experiment 2   -> Tanh + lr=0.01")

print("\nCompare Final Loss Values Above")




print("\nProject Completed Successfully!")
```
